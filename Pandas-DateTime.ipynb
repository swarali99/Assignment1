```python
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p')
df = pd.read_csv('eth.csv', parse_dates=['Date'], date_parser=d_parser)
```

```python
df['DayOfWeek'] = df['Date'].dt.day_name() ## we added the new column of days.
```
